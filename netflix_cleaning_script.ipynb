```python
import pandas as pd
```

```python
#load the dataset
df = pd.read_csv('netflix_titles.csv')
```

```python
#first 5 rows of the dataset
df.head()
```

```python
#check rows and columns
df.shape
```

```python
# 1. Rename columns to be clean and uniform
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.head()
```

```python
# 2. Check and handle missing values
missing_values = df.isnull().sum()
df = df.dropna(subset=['title'])  # Drop rows where title is missing
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['date_added'].fillna('Unknown', inplace=True)
df.head()
```

```python
# 3. Remove duplicate records
df = df.drop_duplicates()
df.shape
```

```python
# 4. Standardize date format
df['date_added'] = df['date_added'].apply(
    lambda x: pd.to_datetime(x, errors='coerce').strftime('%d-%m-%Y') if x != 'Unknown' else x
)
df.head()
```

```python
# 6. Convert duration column
def parse_duration(duration):
    if 'Season' in duration:
        return int(duration.split()[0])
    elif 'min' in duration:
        return int(duration.split()[0])
    return None

df['duration_parsed'] = df['duration'].apply(parse_duration)
df.head()
```

```python

# 7. Save cleaned data
df.to_csv('cleaned_netflix_data.csv', index=False)
```
